```python
import pandas as pd
label_filename = '/content/drive/MyDrive/iCV/new_eeg/train_test_validation/raw/raw_train_class.csv'
ltr = pd.read_csv(label_filename)
ltr
```

```python
df3 = pd.DataFrame()
X_trai = pd.DataFrame()
y_train = pd.DataFrame()
dir = '/content/drive/MyDrive/iCV/new_eeg/train_test_validation/raw/train'
fs = 300
bplist=[]
lblist=[]


for file in os.listdir(dir):
  if file.endswith(".csv"):
    df3 = pd.DataFrame()
    print(file)
    usr = pd.read_csv(dir+'/'+file)
    nt = usr['Time'].iloc[0]
    new_row = usr.loc[(usr['Time'] >= nt+5) & (usr['Time'] <= nt+10)]
    nr = new_row.to_numpy()
    df3 = pd.concat([df3, pd.DataFrame(nr)], ignore_index=True)
    df3.columns = usr.columns
    df4 = df3.drop(df3.columns[0],axis=1)
    a = urlparse(file)
    fn = os.path.basename(a.path)
    #fl = os.path.splitext(fn)[0]
    bplist1=[]
    for column in df4:
      cso = df4[column]
      vals=cso.values
      perm=ant.perm_entropy(vals, order=3, normalize=True)
      svd=ant.svd_entropy(vals, order=3, normalize=True)
      spectral=ant.spectral_entropy(vals, fs, normalize=True)
      app=ant.app_entropy(vals, order=2)
      sample=ant.sample_entropy(vals, order=2)
      petrosian=ant.petrosian_fd(vals)
      katz=ant.katz_fd(vals)
      higuchi=ant.higuchi_fd(vals)
      detrended=ant.detrended_fluctuation(vals)
      hjorth=ant.hjorth_params(vals)
      bplist1.append(perm)
      bplist1.append(svd)
      bplist1.append(spectral)
      bplist1.append(app)
      bplist1.append(sample)
      bplist1.append(petrosian)
      bplist1.append(katz)
      bplist1.append(higuchi)
      bplist1.append(detrended)
      bplist1.append(hjorth[0])
      bplist1.append(hjorth[1])
    bplist1.append(fn)
    bplist.append(bplist1)
    print(len(bplist))
    lbl=ltr.loc[ltr['filename'] == fn, 'class'].iloc[0]
    lblist.append(lbl)
```

```python
bpnp = np.array(bplist)
X_trai = pd.concat([X_trai, pd.DataFrame(bpnp)], ignore_index=True)
```

```python
bpnpl = np.array(lblist)
y_train = pd.concat([y_train, pd.DataFrame(bpnpl, columns=['label'])], ignore_index=True)
```
